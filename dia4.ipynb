## concepts claves de HTTP 
significa HyperText Transfer Protocol es comunicacion entre clientes y un servidor. 
Es un protocolo de solicitud-respuesta para comunicacion entre clientes (navegar, app movil, scrip) y servidores

### Froma general de una peticion 
<METODO> <RUTA> HTTP/1.1
Host: Ejemplo.com
Header-1: valor
Header-2: valor


# metodos HTTP (CRUD)
GET = leer recursos (sin cambiar estado)
POST = crear recursos (cambia estado)
PUT = Reemplazar un recurso completo (actualizar)
PATCH = modifica parcialmente un recurso 
DELETE = borra un recurso 

# idempotencia 
una operacion es idempotente si ejecutarla 1 vez o N vecez produce el mismo estado final del servidor
* idempotentes: GET, PUT, DELETE
* no idempotente: POST (cada post podria crear un nuevo recurso)

## Códigos de estado más comunes
**200 OK**→ todo bien en una lectura/operación.
**201 Created** → recurso creado exitosamente (usar en POST).
204 No Content → operación exitosa sin cuerpo (frecuente en DELETE o PUT).
400 Bad Request → error del cliente (p. ej., JSON mal formado).
401 Unauthorized / 403 Forbidden → falta autenticación o permisos.
404 Not Found → recurso o ruta inexistente.
500 Internal Server Error → error no controlado en el servidor.

### Headers clave
- **Content-Type** formato del cuerpo que **envio** (application/json', application/x-www-form-urlencoded', 'text/html, etc.).
**Accept** formatos que el cliente **acepta recibir**.
**Content-Length cantidad de bytes del cuerpo (el servidor/cliente lo usa para leer el cuerpo).
**Host**, **User-Agent**, **Authorization**, **Cache-Control**, etc. (otros importantes).

# Cuerpo (body) de la petición
* application/json→ envío un JSON (texto) en el cuerpo.
* application/x-www-form-urlencoded → envío pares clave valor&... como en formularios HTML.
* (También existen multipart/form-data para archivos, text/plain, etc.)





# Express
**Express** es un **framework minimalista** para Node.js.
- Simplifica la creación de servidores HTTP, manejando por nosotros tareas comunes como:
- Manejo de rutas (GET, POST, PUT, DELETE).
- Lectura de parámetros y JSON en el body.
- Middleware para procesar datos antes de llegar a la ruta.

## qué es Middleware?
En Express, un middleware es una funcion que se ejecuta entre la peticion request y la respuesta (responde)
**capa intermedia que procesa la peticion o respuesta**

## estructura de un middleware 
```js
function miMiddleware(req, res, next)
{
    console.log("Metodo", req.method, "Ruta", req.url); 
    next(); 
}

app.use(miMiddleware)


```


- `req`:la petición del cliente.
- `res`:la respuesta del servidor.
- `next()`:función que llama al siguiente middleware en la cadena.


## ejemplo de Middleware para procesar json 




````python
const express = require('express');
const app = express(); 
const port = 3000; 

app.use(express.json()); 

app.post("usuario", (req, res) => {
    console.log(req.body); 
    res.json({mensaje: "usuario recibido", datos: req.body}); 
})

app.listen(port, () => { console.log('Servidor escuchando en http://localhost:${port}') });
````

````python
from http.server import BaseHTTPRequestHandler, HTTPServer
import json

class MyHandler(BaseHTTPRequestHandler):
    def send_json(self, data):
        self.send_response(200)
        self.send_header('Content-Type', 'application/json')
        self.end_headers()
        body = json.dumps(data).encode('utf-8')
        self.wfile.write(body)

    def do_GET(self):
        if self.path == '/':
            self.send_json({'message': 'Hola Mundo desde la raiz'})
        elif self.path == '/hello':
            self.send_json({'message': 'Hello World from /hello'})
        else:
            self.send_response(404)
            self.end_headers()
            self.wfile.write(b'Not Found')

if __name__ == '__main__':
    server = HTTPServer(('localhost', 8080), MyHandler)
    print('Servidor HTTP escuchando en http://localhost:8080')
    server.serve_forever()
````

## __dirname, path y middleware en Express 
### Que es __dirname? 
En node.js la variables especial __dirname siempre contiene la ruta absolita de la carpeta en la que se encuentra el archivo actual 

Ejemplo:
Si tienes el archivo `servidor.js` en la carpeta:
- C:\proyectos\mi-servidor-express\servidor.js

El valor de `__dirname` dentro de `servidor.js` será:
- C:\proyectos\mi-servidor-express

### ¿Qué es el módulo path?
El módulo ** path es un módulo interno de Node.js que permite trabajar con rutas de archivos y carpetas de manera
**segura y multiplataforma**.

Por ejemplo:
```js
const path = require("path");

// Une dirname con "public/index.html"
const ruta path.join(__dirname, "public", "index.html");

console.log(ruta);

```

En Windows imprimirá algo como:
C:\proyectos\mi-servidor-express\public/index.html

En Linux/Mac imprimirá:
/home/usuario/mi-servidor-express/public/index.html

## Middleware de archivos estáticos
La línea:
app.use(express.static(path.join(_dirname, "public")));

hace que Express busque automáticamente los archivos en la carpeta public y los sirva directamente al navegador.



# Ejemplo:
* Si existe publiw/index.html, se puede abrir en el navegador con: 
    http://localhost:3000/index.html
* Si existe public/img/logo.png, se abre con: 
    http://localhost:3000/img/logo.png


##Resumen
`_dirname` Ruta absoluta de la carpeta donde está el archivo actual.
`path` Módulo de Node.js para unir y manipular rutas de forma segura.
Middleware en Express Funciones intermedias que procesan las peticiones en orden.
`express.static` Middleware que automáticamente sirve archivos de la carpeta indicada.
